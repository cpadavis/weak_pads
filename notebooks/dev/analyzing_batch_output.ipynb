```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/mbaumer/LSST_work/random_pixel_size/weak_sauce/data/test.csv')
```

```python
feat = df[['Mx','My','flux','e1','e2','inputE','inputS','inputTheta']]#.loc[::4]
```

```python
feat
```

```python
resid_df = pd.DataFrame(columns=df.columns)
for i in np.arange(len(df),step=8):
    resid_df = resid_df.append(df.loc[i]-df.loc[i+4],ignore_index=True) 
    resid_df = resid_df.append(df.loc[i+1]-df.loc[i+5],ignore_index=True) 
    resid_df = resid_df.append(df.loc[i+2]-df.loc[i+6],ignore_index=True) 
    resid_df = resid_df.append(df.loc[i+3]-df.loc[i+7],ignore_index=True) 
```

```python
feat = resid_df[['flux','Mx','My','e1','e2']]
```

```python
downsamp = df.drop(np.arange(len(df),step=2),)
downsamp = downsamp.reset_index()
downsamp[['inputE','inputS','inputTheta']]
```

```python
feat['inputE'] = downsamp['inputE']
feat['inputS'] = downsamp['inputS']
feat['inputTheta'] = downsamp['inputTheta']
```

```python
feat
```

```python
feat
```

```python
arr = []
err_arr = []
for i in np.arange(0,.95,.1):
    subset = feat[np.logical_and(feat.inputE.astype(np.float64) > i-.01, feat.inputE.astype(np.float64) < i+.01) ]
    test = subset.mean()
    test2 = subset.std()
    arr.append([i, test['flux'], test['Mx'], test['My'], test['e1'], test['e2']])
    err_arr.append([i, test2['flux'], test2['Mx'], test2['My'], test2['e1'], test2['e2']])
arr = np.array(arr)
err_arr = np.array(err_arr)
```

```python
arr.shape
```

```python
plt.errorbar(arr[:-4,0],arr[:-4,4],yerr=err_arr[:-4,4]/2)
plt.xlabel('e')
plt.ylabel('e1_real - e1_ideal')
plt.title('Multiplicative Bias (e1)')
```

```python
plt.errorbar(arr[:-4,0],arr[:-4,5],yerr=err_arr[:-4,5]/2)
plt.xlabel('e')
plt.ylabel('e2_real - e2_ideal')
plt.title('Multiplicative Bias (e2)')
```

```python
plt.errorbar(arr[:-3,0],arr[:-3,1],yerr=err_arr[:-3,1]/2)
plt.xlabel('e')
plt.ylabel('flux_real - flux_ideal')
plt.title('Multiplicative Bias')
```

```python
arr = []
err_arr = []
for i in np.arange(3,8.1,1):
    subset = feat[np.logical_and(np.logical_and(feat.inputS.astype(np.float64) > i-.01, feat.inputS.astype(np.float64) < i+.01), feat.inputE == 0 ) ]
    print i, len(subset)
    test = subset.mean()
    test2 = subset.std()
    arr.append([i, test['flux'], test['Mx'], test['My'], test['e1'], test['e2']])
    err_arr.append([i, test2['flux'], test2['Mx'], test2['My'], test2['e1'], test2['e2']])
arr = np.array(arr)
err_arr = np.array(err_arr)
```

```python
plt.errorbar(arr[:,0],arr[:,1]/(.001*np.log(10)/2.5),yerr=err_arr[:,1]/2/(.001*np.log(10)/2.5))
plt.xlabel('fwhm (pixels)')
plt.xlim((2,9))
plt.ylabel('flux_real - flux_ideal (mMag)')
plt.title('Size Bias on Flux')
```

```python
pix_scale = 200
plt.errorbar(arr[:,0],pix_scale*arr[:,2],yerr=pix_scale*err_arr[:,2]/2)
plt.xlabel('fwhm (pixels)')
plt.xlim((2,9))
plt.ylabel('x_real - x_ideal (mas)')
plt.title('Size Bias on astrometry (x-dir)')
```
