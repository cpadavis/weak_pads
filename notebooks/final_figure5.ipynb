```python
import weak_sauce as ws
import weak_sauce.data_tools
from weak_sauce.grid import MoveableGrid
import weak_sauce.shifted_cmap
from scipy import ndimage
```

```python
data_mg = MoveableGrid('/nfs/slac/g/ki/ki19/des/mbaumer/ccd_mg_model_fits/des_chip04_maxit2500_step0.5_decay0.0001/mg.pkl')
```

```python
fig,axs = plt.subplots(1,2,figsize=(5,10),sharey=False,sharex=False)

plt.sca(axs[0])
locs, labels = plt.xticks()
plt.setp(labels, rotation=90)
plt.title('Pixel Widths')

im = axs[0].imshow(ndimage.gaussian_filter(
        (data_mg.source.vertices[:,:,0]-np.roll(data_mg.source.vertices[:,:,0],1,axis=0))[1:-1,1:-1],
        sigma=(2,2), order=0).T,vmin=.988,vmax=1.012,cmap=plt.cm.RdBu_r,origin='lower')
#cbar = plt.colorbar(im,ax=axs[0],ticks=[.99,1,1.01],fraction=0.1)
#cbar.ax.tick_params(labelsize=14) 
axs[0].set_ylabel('Pixel Y')
axs[0].set_xlabel('Pixel X')
axs[1].set_xlabel('Pixel X')
axs[0].set_xticks([200,400,600])
axs[1].set_xticks([200,400,600])

plt.sca(axs[1])
locs, labels = plt.xticks()
plt.setp(labels, rotation=90)
plt.title('Pixel Heights')

im = axs[1].imshow(ndimage.gaussian_filter(
        (data_mg.source.vertices[:,:,1]-np.roll(data_mg.source.vertices[:,:,1],1,axis=1))[1:-1,1:-1],
        sigma=(2,2), order=0).T,vmin=.988,vmax=1.012,cmap=plt.cm.RdBu_r,origin='lower')
cbar = plt.colorbar(im,ax=axs[1],ticks=[.99,1,1.01],fraction=0.1)
cbar.set_label('Relative Pixel Length')
#cbar.set_ticklabels(['-3e-4','0','3e-4'])
#cbar.ax.tick_params(labelsize=14) 
```
